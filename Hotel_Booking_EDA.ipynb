## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
df=hotelbooking_df.copy()
df.drop_duplicates(inplace=True) #Removing Duplicate
````

As we there are 4 variable with Null values from which company and agent are the variable whoes not required to handle because it may be the guest whoes not book room from a company or by any agent. As we see there is some countries also han null value we can remove it by others. And we also sees that there is some null value in children as well and it's very less null values we can remove  it using the mode.

````python
#Handling Nan
df['company'].fillna(0, inplace=True)
df['agent'].fillna(0, inplace=True)
df['country'].fillna("Other", inplace=True)
df['children'].fillna(0, inplace=True)
df.isnull().sum().sort_values(ascending=False)
````

Type Change

Some variable need type change. such as date col which is as object need to convert as datetime format and children in float which is not possible. So we convert children as int64.

````python
df[['children','agent','company']]=df[['children','agent','company']].astype(int)
df['reservation_status_date']=pd.to_datetime(df['reservation_status_date'])
df['arrival_date_year']=df['arrival_date_year'].astype(str)
df.info()
````

````python
df['total_guest']=df['adults']+df['children']+df['babies']
df['total_stay']=df['stays_in_week_nights']+df['stays_in_weekend_nights']
df.drop(df[df['total_guest']==0].index, inplace= True)
````

### What all manipulations have you done and insights you found?

I removed duplicates form the data. i replaced nan of company and agent from 0 which repesnt as blank value. i also replace nan of country variable with others. i replaced nam of children with 0 which is mode of the children . i added a col of total guest which is the of adult + children + babies. There are some rows available where total guest is 0 so i removed these rows.i Also added a column named total_stay which is sum of 'stays_in_week_nights' & 'stays_in_weekend_nights'.
